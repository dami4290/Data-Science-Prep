```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("https://dlsun.github.io/stats112/data/titanic.csv")
```

```python
crosstab=df[["pclass", "survived"]].value_counts().unstack("pclass")
survived_totals = crosstab.sum(axis="columns")
crosstab.divide(survived_totals, axis="rows").plot.bar(stacked=True)
```

```python
crosstab=df[["sex", "survived"]].value_counts().unstack("sex")
survived_totals = crosstab.sum(axis="columns")
crosstab.divide(survived_totals, axis="rows").plot.bar(stacked=True)
```

```python
crosstab=df[["age", "survived"]].value_counts().unstack("age")
survived_totals = crosstab.sum(axis="columns")
crosstab.divide(survived_totals, axis="rows").plot.bar(stacked=True)
```

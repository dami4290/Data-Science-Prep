```python
nyu["Foreign Gift Amount"]=nyu["Foreign Gift Amount"].str.replace("$","")
nyu.head()
```

```python
nyu["Foreign Gift Amount"]=nyu["Foreign Gift Amount"].str.replace(",","")
nyu.head()
```

```python
nyu["amount_numeric"]=nyu["Foreign Gift Amount"].astype("float")
nyu.head()
```

```python
nyu.sort_values(by=["amount_numeric"],ascending=False,inplace=True)
nyu.head()
```

```python
nyu["amount_numeric"].mean()
nyu.describe()
```
